From the sample of the dataset above, it can be seen that some symbols that need to be removed amongst other preprocessing. Let's write some functions that will take care of the preprocessing.

```python
import contractions
import re
from nltk.corpus import stopwords

stop_words = stopwords.words('english') 
stop_words.remove('not') # since "not" affects the sentiment of the review.

def fix_contractions(text):
    return contractions.fix(text)
    
def remove_punctuations(text):
    return re.sub('[^a-zA-Z]', ' ', text)

def convert_to_lower(text):
    return text.lower()
 
def remove_stopwords(text):
    return  ' '.join(word for word in text.split() if word not in stop_words) 

def preprocess(text):
    text = fix_contractions(text)
    text = remove_punctuations(text)
    text = convert_to_lower(text)
    text = remove_stopwords(text)
    return text

cleaned_data['Review'] = cleaned_data['Review'].apply(preprocess)
```

```python
cleaned_data['Review']
```

We have been able to clean our data. Let's move on to changing our preprocessed data into a usable format.

## Feature Engineering

We'll need to convert our now preprocessed text into usable features that can be processed by the selected algorithms.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
```

```python
vectorized_data = vectorizer.fit_transform(cleaned_data['Review']).toarray() # sparse matrix to array
vectorized_data.shape
```

```python
vectorizer.vocabulary_
```

```python
vectorized_data
```

## Model Creation and Evaluation

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(vectorized_data, cleaned_data['Liked'], test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import precision_score, recall_score

sgd = SGDClassifier()
sgd.fit(X_train, y_train)

y_pred = sgd.predict(X_test)
```

```python
precision_score(y_true=y_test,
                y_pred=y_pred).round(3)
```

```python
recall_score(y_true=y_test,
             y_pred=y_pred).round(3)
```

Adjusting the threshold of the classifier will give you more control over the precision and recall values of the classifier.

### Visualizing the Threshold

```python
from sklearn.metrics import precision_recall_curve
from sklearn.model_selection import cross_val_predict

predictions = cross_val_predict(estimator=sgd,
                                X=X_train,
                                y=y_train,
                                cv=3,
                                method='decision_function')
```

```python
precisions, recalls, thresholds = precision_recall_curve(y_train, predictions)

threshold_index = (precisions >= recalls).argmax()
threshold = thresholds[threshold_index]

def plot_pr_comparison_curve(thresholds, precisions, recalls, threshold_value=0):
    
    plt.figure(figsize=(8, 4)) 
    plt.grid()
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
    plt.vlines(threshold_value, 0, 1.0, "k", "dotted", label="threshold")
    plt.xlabel("Threshold")
    plt.legend(loc="center right")
    
    plt.show()

plot_pr_comparison_curve(thresholds=thresholds,
              precisions=precisions,
              recalls=recalls,
              threshold_value = threshold)
              
```

Let's adjust the threshold to achieve a recall of about 90%.

```python
recall_at_90_index = (recalls <= 0.9).argmax() # <= is used instead of >= since the recall values are in descending order.

recall_90_threshold = thresholds[recall_at_90_index]
```

```python
predictions_at_90_recall = (predictions >= recall_90_threshold).astype(int)


recall_score(y_true=y_train,
             y_pred=predictions_at_90_recall).round(2)
```

```python
precision_score(y_true=y_train,
                y_pred=predictions_at_90_recall).round(2)
```

We have succesfully adjusted our threshold to give us about 90% recall but at the cost of a reduced precision since there's a tradeoff between precision and recall, as can be noted from the plot above. But having a recall of about 90% and precision at 70% is still quite good depending on the learning task.

We can also do the same for precision. Let's adjust the threshold to a precision of about 95%.

```python
precision_at_95_index = (precisions >= 0.95).argmax()
precision_95_threshold = thresholds[precision_at_95_index]
```

```python
predictions_at_95_precision = (predictions >= precision_95_threshold).astype(int)

precision_score(y_true=y_train,
                y_pred=predictions_at_95_precision).round(3)
```

```python
recall_score(y_true=y_train,
                y_pred=predictions_at_95_precision).round(3)
```

Let's visualize this

```python
plt.figure(figsize=(8, 4)) 
plt.grid()
plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
plt.plot(precision_95_threshold, precisions[precision_at_95_index], "bo")
plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
plt.plot(precision_95_threshold, recalls[precision_at_95_index], "go")
plt.vlines(precision_95_threshold, 0, 1.0, "k", "dotted", label="threshold")
plt.xlabel("Threshold")
plt.legend(loc="center right")

plt.show()
```

Getting the classifier's precision at 95% will lead to a recall of 39% which doesn't make for a very good classifier.

### Evaluation using Precision Recall Curve

```python
def plot_pr_curve(recalls, precisions):
    plt.figure(figsize=(6, 5))
    plt.plot(recalls, precisions, linewidth=2, label="Precision/Recall curve")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    #plt.axis([0, 1, 0, 1])
    plt.grid()
    plt.legend(loc="lower left")
    
    plt.show()

plot_pr_curve(recalls=recalls,
              precisions=precisions)
```

### Evaluating Using the ROC Curve

```python
from sklearn.metrics import roc_curve

false_positive_rate, true_positive_rate, thresholds = roc_curve(y_true=y_train,
                                                               y_score=predictions)
```

```python
def plot_roc_curve(false_positive_rate, true_positive_rate):
    plt.figure(figsize=(6, 5))
    plt.plot(false_positive_rate, true_positive_rate, linewidth=2, label="ROC curve")
    plt.plot([0, 1], [0, 1], 'k:', label="Random classifier's ROC curve")
    
    
    
    plt.xlabel('False Positive Rate (Fall-Out)')
    plt.ylabel('True Positive Rate (Recall)')
    plt.grid()
    plt.axis([0, 1, 0, 1])
    plt.legend(loc="lower right", fontsize=13)
    
    plt.show()
    
plot_roc_curve(false_positive_rate=false_positive_rate,
               true_positive_rate=true_positive_rate)
```

From the PR-curve and the ROC curve above, it can be seen that the classifier has more room for improvement. Let's scale the features to see if there'll be an improvment.

### Error Analysis.

Now let's try and find out some of the mistakes our selected model makes and try to fix to get better results. We'll do this by using a confusion matrix.

```python
from sklearn.metrics import ConfusionMatrixDisplay

y_train_pred = cross_val_predict(log_clf, X_train, y_train, cv=3)
plt.rc('font', size=9) 
ConfusionMatrixDisplay.from_predictions(y_true=y_train,
                                        y_pred=y_train_pred,
                                        normalize="true",
                                        values_format=".0%")
plt.show()
```

From above it can be seen that our chose classifier has a false negative rate of 19% and a false positive rate of 18%.
Let's try to fix this by using the OVO classification method.

```python
from sklearn.multiclass import OneVsOneClassifier

ovo_log_clf = OneVsOneClassifier(SGDClassifier(loss='log_loss'))

ovo_log_clf.fit(X_train, y_train)

ovo_log_predictions = ovo_log_clf.predict(X_test)

```

```python
precision_score(y_true=y_test,
                y_pred=ovo_log_predictions)
```

```python
recall_score(y_true=y_test,
             y_pred=ovo_log_predictions)
```

```python
f1_score(y_true=y_test,
         y_pred=ovo_log_predictions)
```

```python
ovo_y_train_pred = cross_val_predict(ovo_log_clf, X_train, y_train, cv=3)
plt.rc('font', size=9) 
ConfusionMatrixDisplay.from_predictions(y_true=y_train,
                                        y_pred=ovo_y_train_pred,
                                        normalize="true",
                                        values_format=".0%")
plt.show()
```
